```python
import pandas as pd
df=pd.read_excel("/content/drive/MyDrive/datasetvoiture.xlsx")
df
```

```python
df.drop('matricule', axis=1, errors='ignore',inplace=True)
df
```

```python
df[['marque']]
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
ohe.fit(df[['marque']])
ohe.categories_[0]
```

```python
marque_ohe=ohe.transform(df[['marque']])
marque_ohe.toarray()
```

```python
df_marque=pd.DataFrame(marque_ohe.toarray(), columns='marque_'+ohe.categories_[0])
df_marque
```

```python
df=pd.concat([df,df_marque], axis=1)
df
```
